## Merge all csv files to one

```python
import pandas as pd
import glob
import os

hashtag = 'pastelcartel'

files = os.path.join("./data/TikTok_pyktok/metadata/{}/".format(hashtag), "*.csv")
files = glob.glob(files)

# joining files with concat and read_csv
df = pd.concat(map(pd.read_csv, files), ignore_index=True)
df.to_csv('./data/Tiktok_pyktok/merged_metadata/{}_tiktok_pyktok_metadata.csv'.format(hashtag),
         encoding='utf-8-sig')
```

### escobars

```python
#escobars

#use 'video_description' column

escobars = pd.read_csv('./data/Tiktok_pyktok/merged_metadata/escobars_tiktok_pyktok_metadata.csv')
escobars
```

```python
escobars.info()
```

```python
#langdetect - https://pypi.org/project/langdetect/

from langdetect import detect
escobars['video_description'] = escobars['video_description'].astype(str)
escobars['video_description_preprocessing'] = escobars['video_description'].apply(detect)
escobars = escobars[(escobars['video_description_preprocessing'] == 'en') | (escobars['video_description_preprocessing'] == 'es')]
```

```python
#delete '#'
#stopwords: 'fyp'..

escobars['video_description'] = escobars['video_description'].str.replace('#', ' ')
```

```python
from itertools import chain
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt
%matplotlib inline


escobars_new = escobars[escobars['video_description'].notnull()]

docs_words = [x.split(' ') for x in escobars_new['video_description']]

total_docs_words = []
for words in docs_words:
    total_docs_words.extend(words)
    
    
stopwords = ['the','to','of','in','is','and','fyp', 'india', 'tiktok',
            'are', 'me', 'this', 'my', 'fyp', 'foryou', 'foryoupage', 'you', 'so', 'it', 'on', 'do', 'about',
            'her', 'be', 'fyp', 'when', 'your', 'he', 'our', 'by', 'we','video', 'for', 'viral', 'at','with', 'trending',
            'april', 'may', 'like', 'fyp„Ç∑', '', 'by:', 'follow', 'us!', 'escobars', 'Escobars', 'escobar',
            '@LdC1']

data_map = Counter(x for x in total_docs_words if x not in stopwords and len(x) != 1)
data_map.most_common(n=30)
```

```python
len(data_map)
```

```python
#create wordcloud
path = (r'/Users/sungwonj/Downloads/NanumGothicExtraBold.otf') #font path
wordcloud = WordCloud(font_path=path, max_font_size=80, background_color='white', max_words= 100)
wc = wordcloud.generate_from_frequencies(data_map)
fig = plt.figure(figsize = (10,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
import csv

with open('./data/TikTok_escobars_word_frequencies.csv','w',
         encoding='utf-8-sig') as csvfile:
    writer=csv.writer(csvfile)
    writer.writerows(data_map.items())
```

```python
escobars = pd.read_csv('./data/TikTok_escobars_word_frequencies.csv', header=None)
escobars.columns = ['word', 'frequency']
```
